```python
import pandas as pd

jan_sleep_data = pd.read_csv('/Users/ruojiakuang/Desktop/GSB 545 Advanced ML/Lab 1/January sleep data - Sheet1.csv')
feb_sleep_data = pd.read_csv('/Users/ruojiakuang/Desktop/GSB 545 Advanced ML/Lab 1/February sleep data - Sheet1.csv')
mar_sleep_data = pd.read_csv('/Users/ruojiakuang/Desktop/GSB 545 Advanced ML/Lab 1/March sleep data - Sheet1.csv')
apr_sleep_data = pd.read_csv('/Users/ruojiakuang/Desktop/GSB 545 Advanced ML/Lab 1/April sleep data - Sheet1.csv')
nov_sleep_data = pd.read_csv('/Users/ruojiakuang/Desktop/GSB 545 Advanced ML/Lab 1/November Sleep Data - Sheet1.csv')
dec_sleep_data = pd.read_csv('/Users/ruojiakuang/Desktop/GSB 545 Advanced ML/Lab 1/December Sleep data - Sheet1.csv')
```

```python
# Rename the column to 'Days of the week', and add a new column 'Month'
jan_sleep_data = jan_sleep_data.rename(columns={'JANUARY': 'DAYS OF THE WEEK'})
jan_sleep_data['Month'] = 'January'

feb_sleep_data = jan_sleep_data.rename(columns={'February': 'DAYS OF THE WEEK'})
feb_sleep_data['Month'] = 'February'

mar_sleep_data = jan_sleep_data.rename(columns={'March': 'DAYS OF THE WEEK'})
mar_sleep_data['Month'] = 'March'

apr_sleep_data = jan_sleep_data.rename(columns={'April': 'DAYS OF THE WEEK'})
apr_sleep_data['Month'] = 'April'

nov_sleep_data = jan_sleep_data.rename(columns={'November': 'DAYS OF THE WEEK'})
nov_sleep_data['Month'] = 'November'

dec_sleep_data = jan_sleep_data.rename(columns={'December': 'DAYS OF THE WEEK'})
dec_sleep_data['Month'] = 'December'
```

```python
# Concatenate the DataFrames
sleep_data = pd.concat([jan_sleep_data, feb_sleep_data, mar_sleep_data, apr_sleep_data, nov_sleep_data, dec_sleep_data])
sleep_data
```

```python
sleep_data_cleaned = sleep_data.dropna()
```

```python
# Convert 'SLEEP TIME' to 'SLEEP START' and 'SLEEP END' in hours
sleep_data_cleaned[['SLEEP START', 'SLEEP END']] = sleep_data_cleaned['SLEEP TIME'].str.split(' - ', expand=True)

sleep_data_cleaned['SLEEP START'] = pd.to_datetime(sleep_data_cleaned['SLEEP START']).dt.hour
sleep_data_cleaned['SLEEP END'] = pd.to_datetime(sleep_data_cleaned['SLEEP END']).dt.hour

sleep_data_cleaned = sleep_data_cleaned.drop(['SLEEP TIME'], axis=1)

# Convert 'HOURS OF SLEEP' to numeric numbers
sleep_data_cleaned['HOURS OF SLEEP'] = sleep_data_cleaned['HOURS OF SLEEP'].str.split(':').apply(lambda x: int(x[0]) + int(x[1])/60)

# Convert percentage columns to numeric
sleep_data_cleaned['REM SLEEP'] = sleep_data_cleaned['REM SLEEP'].str.rstrip('%').astype('float') / 100
sleep_data_cleaned['DEEP SLEEP'] = sleep_data_cleaned['DEEP SLEEP'].str.rstrip('%').astype('float') / 100
sleep_data_cleaned['HEART RATE UNDER RESTING'] = sleep_data_cleaned['HEART RATE UNDER RESTING'].str.rstrip('%').astype('float') / 100

# Encode categorical variables
sleep_data_cleaned = pd.get_dummies(sleep_data_cleaned, columns=['DAYS OF THE WEEK', 'Month'])
```

```python
# Split dataset into features and labels
from sklearn.model_selection import train_test_split
X = sleep_data_cleaned[['HOURS OF SLEEP', 'REM SLEEP', 'DEEP SLEEP', 'HEART RATE UNDER RESTING', 'SLEEP START', 'SLEEP END', 'DAYS OF THE WEEK_Friday', 'DAYS OF THE WEEK_Monday', 'DAYS OF THE WEEK_Saturday', 'DAYS OF THE WEEK_Sunday', 'DAYS OF THE WEEK_Thursday', 'DAYS OF THE WEEK_Tuesday', 'DAYS OF THE WEEK_Wednesday', 'Month_April', 'Month_December', 'Month_February', 'Month_January', 'Month_March', 'Month_November']]
y = sleep_data_cleaned['SLEEP SCORE']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5) # 70% training and 30% test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import StackingRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

# Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf=rf_model.predict(X_test)
```

```python
# Stacking Model
estimators = [
    ('rf', RandomForestRegressor(n_estimators=100, random_state=42)),
    ('lr', LinearRegression()),
    ('dt', DecisionTreeRegressor())
]

stack_reg = StackingRegressor(estimators=estimators, final_estimator=LinearRegression())
stack_reg.fit(X_train, y_train)
y_pred_stack = stack_reg.predict(X_test)
```

```python
from sklearn.metrics import *

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
print(f'Random Forest - Mean Squared Error: {mse_rf}, R² Score: {r2_rf}, MAE: {mae_rf}')

mse_stack = mean_squared_error(y_test, y_pred_stack)
r2_stack = r2_score(y_test, y_pred_stack)
mae_stack = mean_absolute_error(y_test, y_pred_stack)
print(f'Stacking Regressor - Mean Squared Error: {mse_stack}, R² Score: {r2_stack},  MAE: {mae_stack}')
```

```python
# Define parameter grid for grid search
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 20]
}
```

```python
rf_model = RandomForestRegressor(random_state=42)
grid_search_rf = GridSearchCV(rf_model, param_grid_rf, cv=5, scoring='neg_mean_squared_error')
grid_search_rf.fit(X_train, y_train)

best_params_rf = grid_search_rf.best_params_
best_estimator_rf = grid_search_rf.best_estimator_

best_estimator_rf.fit(X_train, y_train)

y_pred_rf = best_estimator_rf.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)

print("Best parameters for RandomForestRegressor:", best_params_rf)
print("Random Forest - Mean Squared Error:", mse_rf)
print("Random Forest - R² Score:", r2_rf)
print("Random Forest - MAE", mae_rf)
```

```python
rf_params = {
    'n_estimators': 300,
    'max_depth': 10,
}
rf_model = RandomForestRegressor(**rf_params)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
mae = mean_absolute_error(y_test, y_pred_rf)

print("Random Forest - MAE Score:", mae)
print("Random Forest - Mean Squared Error:", mse_rf)
print("Random Forest - R² Score:", r2_rf)
```

```python
estimators = [
    ('rf', RandomForestRegressor(random_state=42)),
    ('lr', LinearRegression()),
    ('dt', DecisionTreeRegressor(random_state=42))
]

final_estimator = LinearRegression()

param_grid_stack = {
    'rf__n_estimators': [50, 100, 200],
    'rf__max_depth': [None, 10, 20],
    'dt__max_depth': [None, 10, 20],
    'dt__min_samples_split': [2, 5, 10],
    'dt__min_samples_leaf': [1, 2, 4],
}


stack_model = StackingRegressor(estimators=estimators, final_estimator=final_estimator)
grid_search_stack = GridSearchCV(stack_model, param_grid_stack, cv=5, scoring='neg_mean_squared_error')
grid_search_stack.fit(X_train, y_train)

best_params_stack = grid_search_stack.best_params_
best_estimator_stack = grid_search_stack.best_estimator_

best_estimator_stack.fit(X_train, y_train)
y_pred_stack = best_estimator_stack.predict(X_test)

mse_stack = mean_squared_error(y_test, y_pred_stack)
r2_stack = r2_score(y_test, y_pred_stack)
mae_stack = mean_absolute_error(y_test, y_pred_stack)

print("Best parameters for StackingRegressor:", best_params_stack)
print("Stacking Regressor - Mean Squared Error:", mse_stack)
print("Stacking Regressor - R² Score:", r2_stack)
print("Stacking Regressor - MAE", mae_stack)
```
